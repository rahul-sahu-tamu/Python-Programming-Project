```python
import pandas as pd
from pdf2image import convert_from_path
import pytesseract
from pytesseract import image_to_string
from PIL import Image
pytesseract.pytesseract.tesseract_cmd = "C:/Users/R-Sahu/AppData/Local/Programs/Tesseract-OCR/tesseract.exe"
```

```python
def text_extractor(pdf):
    images=convert_from_path(pdf)
    text=""
    for pg, img in enumerate(images):
        text+=image_to_string(img)
    return text
```

```python
crash_data=pd.read_csv("crash_escooter.csv")
#crash_data["e-scooter"]=0
crash_data["pdf_text"]=str(0)
crash_data["crash_id"]=crash_data["crash_id"].astype(str)
#keywords=["scooter","bird","lime","wheels","lyft"]

for i in range(len(crash_data["crash_id"])):
    pdf_location="C:/Users/R-Sahu/Documents/RS/E-Scooter Crash Reports/"+crash_data["crash_id"][i]+".pdf"
    print("Reading "+str(i)+" "+crash_data["crash_id"][i])
    pdf_text=text_extractor(pdf_location)
    pdf_text_lower=pdf_text.lower()
    crash_data["pdf_text"][i]=pdf_text_lower
    #if any(word in pdf_text_lower for word in keywords):
    #    crash_data["e-scooter"][i]=1
    #    print("Scooter Crash found in "+crash_data["crash_id"][i])
print("Extraction Complete")
```
